```python
import matplotlib.pyplot as plt
import numpy as np

def model_one():
    x = np.arange(0.0,2.0, 0.01)
    y = np.sin(2*np.pi*x)
    plt.plot(x,y,"o")
    plt.xlabel('time (s)')
    plt.ylabel('volts (mV)')
    plt.show()
```

```python
model_one()
```

```python
import matplotlib.pyplot as plt
import random

def model_two(npts):
    numbers = []
    for _ in range(npts):
        numbers.append(random.random())
    plt.hist(numbers, 10)
    plt.show()

if __name__ == "__main__":
    model_two(100)
```
